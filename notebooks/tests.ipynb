```python
import pandas as pd
import numpy as np

# Ścieżka do pliku
raw_file_path = 'D:/BK_dokumenty_D2/LendingClub/accepted_2007_to_2018Q4.csv'
```

```python
# Wybieram potrzebne kolumny

# (bez nrows),tylko wybrane kolumny
df = pd.read_csv(raw_file_path, low_memory=False)
```

```python
summary = df.describe(include="all").T
summary
```

```python
pd.set_option("display.max_rows", None)
display(summary)
pd.reset_option("display.max_rows")
```

```python
summary[summary['unique'] < 10]
```

```python
cols_small_unique = summary[summary['unique'] < 10].index.tolist()
for col in cols_small_unique:
    print(f"Column: {col}")
    print(df[col].value_counts(dropna=False))  # includes NaN if any
    print("-" * 40)
```

```python
from IPython.display import display, HTML

output = ""
for col in cols_small_unique:
    output += f"<b>Column: {col}</b><br>"
    output += df[col].value_counts(dropna=False).to_frame().to_html()
    output += "<hr>"

display(HTML(f"<div style='height:600px; overflow:auto'>{output}</div>"))
```

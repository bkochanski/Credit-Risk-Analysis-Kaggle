```python
import pandas as pd
import numpy as np

def describe_with_nans(df):
    # Basic describe, transposed for readability
    desc = df.describe(include="all").T
    
    # Add dtype
    desc["dtype"] = df.dtypes
    
    # Add missing info
    desc["missing"] = df.isna().sum()
    desc["missing_ratio"] = df.isna().mean()
    
    # For numeric columns, compute median ignoring NaNs
    numeric_cols = df.select_dtypes(include=np.number).columns
    desc.loc[numeric_cols, "median"] = df[numeric_cols].median()
    
    return desc
summary = describe_with_nans(df)
summary
```

```python
import pandas as pd
import numpy as np

def optimize_and_save_df(df, parquet_path=None, category_thresh=50, compression="brotli"):
    """
    Optimizes a pandas DataFrame:
    - Converts object columns with <= category_thresh unique values to category
    - Safely downcasts numeric columns (int64 -> int8/int16/int32, float64 -> float32)
    - Optionally saves to compressed Parquet
    
    Args:
        df (pd.DataFrame): Input DataFrame
        parquet_path (str, optional): Path to save Parquet. If None, does not save
        category_thresh (int): Max unique values to convert object -> category
        compression (str): Parquet compression ("gzip", "brotli", "zstd")
    
    Returns:
        pd.DataFrame: Optimized DataFrame
    """
    
    df_optimized = df.copy()
    
    print("Memory usage before optimization:")
    print(df_optimized.info(memory_usage="deep"))
    
    # --------------------------
    # 1️⃣ Convert low-cardinality objects to category
    # --------------------------
    obj_cols = df_optimized.select_dtypes(include="object").columns
    for col in obj_cols:
        n_unique = df_optimized[col].nunique(dropna=True)
        if n_unique <= category_thresh:
            df_optimized[col] = df_optimized[col].astype("category")
    
    # --------------------------
    # 2️⃣ Downcast integers safely
    # --------------------------
    int_cols = df_optimized.select_dtypes(include="int64").columns
    for col in int_cols:
        col_min, col_max = df_optimized[col].min(), df_optimized[col].max()
        if -128 <= col_min and col_max <= 127:
            df_optimized[col] = df_optimized[col].astype("int8")
        elif -32768 <= col_min and col_max <= 32767:
            df_optimized[col] = df_optimized[col].astype("int16")
        elif -2147483648 <= col_min and col_max <= 2147483647:
            df_optimized[col] = df_optimized[col].astype("int32")
        # else keep int64
    
    # --------------------------
    # 3️⃣ Downcast floats safely
    # --------------------------
    float_cols = df_optimized.select_dtypes(include="float64").columns
    for col in float_cols:
        df_optimized[col] = df_optimized[col].astype("float32")
    
    print("\nMemory usage after optimization:")
    print(df_optimized.info(memory_usage="deep"))
    
    # --------------------------
    # 4️⃣ Optional: Save to Parquet
    # --------------------------
    if parquet_path:
        df_optimized.to_parquet(parquet_path, index=False, compression=compression)
        print(f"\nDataFrame saved to {parquet_path} with {compression} compression")
    
    return df_optimized
```

```python
import pandas as pd
import numpy as np

def full_describe(df):
    """
    Enhanced descriptive statistics for numeric, datetime, and categorical columns.
    - Numeric & datetime treated the same (mean, std, quantiles, min/max).
    - Categorical returns top 3 categories with counts + min/median/max.
    """

    summary_list = []

    for col in df.columns:
        s = df[col]
        dtype = s.dtype
        non_na = s.dropna()

        desc = {
            "dtype": dtype,
            "count": s.count(),
            "missing": s.isna().sum(),
            "missing_ratio": s.isna().mean(),
            "mean": np.nan,
            "std": np.nan,
            "min": np.nan,
            "25%": np.nan,
            "50%": np.nan,
            "75%": np.nan,
            "max": np.nan,
            "median": np.nan,
            "unique": np.nan,
            "top_1": np.nan, "top_1_freq": np.nan,
            "top_2": np.nan, "top_2_freq": np.nan,
            "top_3": np.nan, "top_3_freq": np.nan,
        }

        # ----------------------------------------------------------
        # Numeric and datetime: treat the same
        # ----------------------------------------------------------
        if pd.api.types.is_numeric_dtype(s) or pd.api.types.is_datetime64_any_dtype(s):
            if len(non_na) > 0:
                desc.update({
                    "mean": non_na.mean(),
                    "std": non_na.std(),
                    "min": non_na.min(),
                    "25%": non_na.quantile(0.25),
                    "50%": non_na.quantile(0.50),
                    "median": non_na.quantile(0.50),
                    "75%": non_na.quantile(0.75),
                    "max": non_na.max(),
                })

        # ----------------------------------------------------------
        # Categorical / object columns
        # ----------------------------------------------------------
        else:
            # unique count
            desc["unique"] = non_na.nunique()

            # top 3 categories
            value_counts = non_na.value_counts()
            for i in range(min(3, len(value_counts))):
                desc[f"top_{i+1}"] = value_counts.index[i]
                desc[f"top_{i+1}_freq"] = value_counts.iloc[i]

            # lexicographical min/median/max
            try:
                sorted_vals = sorted(non_na.astype(str))
                if sorted_vals:
                    desc["min"] = sorted_vals[0]
                    desc["max"] = sorted_vals[-1]
                    desc["median"] = sorted_vals[len(sorted_vals)//2]
            except:
                pass

        summary_list.append((col, desc))

    return pd.DataFrame.from_dict(dict(summary_list), orient="index")


summary2 = full_describe(df)
summary2
```
